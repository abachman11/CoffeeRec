```python
from pymongo import MongoClient
from prettytable import PrettyTable
from bson.code import Code


mongo = MongoClient('localhost', 3001)
db = mongo.meteor
tweets = db.tweets
```

Find the number of tweets for each coffee

```python
coffee_counts = tweets.aggregate([
            {"$group": {"_id": "$coffee_name", "count": {"$sum": 1}}},
        ])
pt = PrettyTable(field_names=['Coffee Name', 'Count'])
[ pt.add_row((row['_id'], row['count'])) for row in coffee_counts['result'] ]
pt.max_width['Text'] = 50
pt.align= 'l'
print pt

```

```python
mapper = Code("""
    function() {
        emit(this.tweet.user.screen_name, this.coffee_name);
    }
""")

reducer = Code("""
    function(key, values) {
      var total = "";
      for (var i = 0; i < values.length; i++) {
        total += values[i];
      }
      return total;
    }
""")

result = tweets.map_reduce(mapper, reducer, "myresults")

pt = PrettyTable(field_names=['Username', 'Coffees'])
[ pt.add_row((row['_id'], row['value'])) for row in result.find() ]
pt.max_width['Text'] = 50
pt.align= 'l'
print pt
```
