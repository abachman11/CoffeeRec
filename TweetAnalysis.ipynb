```python
from pymongo import MongoClient
from prettytable import PrettyTable
from bson.son import SON


mongo = MongoClient('localhost', 3001)
db = mongo.meteor
tweets = db.tweets
```

# Example Document Structure

```python
tweet = tweets.find_one()
print "roaster name: ", tweet['roaster_name']
print "coffee name: ", tweet['coffee_name']
import json

print "tweet :", json.dumps(tweet['tweet'], sort_keys=True, indent=4, separators=(',', ': '))
```

```python
tweets_summary = [
    (row['coffee_name'], row['roaster_name'], row['tweet']['text'])
    for row in tweets.find()
]
pt = PrettyTable(field_names=['Coffee Name', 'Roaster Name', 'Content'])
[ pt.add_row(row) for row in tweets_summary ]
pt.max_width['Content'] = 20
pt.align= 'l'
print pt


```

###Validity of the Tweets
From the table generated above I counted the number of tweets that correctly identified the coffee that was referenced in the tweets. 168 out of 190 tweets that were counted correctly identified the coffee. 
The incorrect coffees were almost all refering to Starbucks coffees. A large number of the incorrectly identified tweets were due to the names of the coffees being country names, so Starbucks opening a store in Puerto Rico caused a large number of tweets identified as being part of the 

```python
coffee_counts = tweets.aggregate([
         {"$group": {"_id": "$coffee_name", "count": {"$sum": 1}}},
         {"$sort": SON([("count", -1), ("_id", -1)])}
     ])

pt = PrettyTable(field_names=['Coffee Name', 'Count'])
[ pt.add_row((row['_id'], row['count'])) for row in coffee_counts['result'] ]
pt.max_width['Text'] = 50
pt.align= 'l'
print pt
```

```python
roaster_counts = tweets.aggregate([
         {"$group": {"_id": "$roaster_name", "count": {"$sum": 1}}},
         {"$sort": SON([("count", -1), ("_id", -1)])}
     ])

pt = PrettyTable(field_names=['Roaster Name', 'Count'])
[ pt.add_row((row['_id'], row['count'])) for row in roaster_counts['result'] ]
pt.max_width['Text'] = 50
pt.align= 'l'
print pt
```

```python
roaster_counts = tweets.aggregate([
         {"$group": {"_id": "$tweet.user.screen_name", "count": {"$sum": 1}, "roasters": {"$addToSet": "$roaster_name" }}},
         {"$sort": SON([("count", -1), ("_id", -1)])}
     ])
pt = PrettyTable(field_names=['Roaster Name', 'Count', "Screen Names"])
[ pt.add_row((row['_id'], row['count'], row['roasters'])) for row in roaster_counts['result'] ]
pt.max_width['Text'] = 50
pt.align= 'l'
print pt
```
